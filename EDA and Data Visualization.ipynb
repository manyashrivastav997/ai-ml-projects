```python
# Univariate Analysis (Numerical)
# Histogram for Age(Matplotlib)
plt.hist(df['Age'], bins=5, color='lightblue', edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Observation:
The histogram shows that most individuals are between 20 and 60 years old.  
However, the age range 25 to 35 has slightly fewer entries compared to other groups which  
indicates a small dip in that age segment.

```python
# Univariate Analysis (Categorical)
# Workout Type Distribution (Pie Chart)
count= df['Workout_Type'].value_counts()
count
plt.pie(count,labels= count.index,autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title("Workout Type Distribution")
plt.show()
```

Observation:
Most people are evenly distributed among different workout types, that are Yoga, Strength, HIIT and Cardio, with only small differences in their proportions. This shows that all workout types are almost equally preferred.

```python
# Univariate Analysis (Numerical)
# Box Plot for Height(Seaborn)
sns.boxplot(x=df['Height (m)'], color='purple')
plt.title("Height Distribution (Box Plot)")
plt.xlabel("Height (m)")
plt.show()
```

The box plot shows that most individuals have a height between 1.6 m and 1.8 m, with the median (middle value) around 1.7 m, indicating that the height distribution is fairly balanced.
The whiskers extend slightly on both sides, showing a moderate spread of data.
There are no major outliers, meaning most people’s heights fall within a normal range without any extremely short or tall individuals.

```python
#Bivariate Analysis (Both Numerical)
#Height vs Weight (Scatter plot using Matplotlib)
plt.figure(figsize=(8,5))
plt.scatter(df['Height (m)'], df['Weight (kg)'], color='green', edgecolor='white', s=60, alpha=0.6)
plt.title("Height vs Weight Relationship", fontsize=13)
plt.xlabel("Height (m)")
plt.ylabel("Weight (kg)")
plt.grid(True, linestyle='--')
plt.show()
```

The scatter plot shows that as height increases, weight also increases for most people. This means taller people usually weigh more . It is because the points are spread in an upward pattern, showing a positive relationship between height and weight.

```python
# Bivariate Analysis (Both Categorical)
# Workout Type vs Burns_Calories_Bin using Seaborn Countplot
plt.figure(figsize=(8,5))
sns.countplot(x='Workout_Type',hue='Burns_Calories_Bin', data=df, palette='Set1')
plt.title("Workout Type vs Calories Burned Category")
plt.xlabel("Workout Type")
plt.ylabel("Count")
plt.legend(title="Calories Burned")
plt.show()
```

Most workout types including Strength, HIIT, Cardio, and Yoga, show a fairly balanced distribution across all calorie burn categories (Low, Medium, High and Very High). This means that each workout type contributes almost equally to different calorie burn levels

```python

#Bivariate Analysis(One numerical and one categorical)
# Boxplot: Calories_Burned vs Workout_Type using Seaborn
sns.boxplot(x='Workout_Type', y='Calories_Burned', data=df,hue='Burns_Calories_Bin', palette='Set2')
plt.title("Calories Burned Across Different Workout Types")
plt.xlabel("Workout Type")
plt.ylabel("Calories Burned")
plt.show()
```

Observation: The boxplot shows that different workout types (Strength, HIIT, Cardio, Yoga) have a wide range of calories burned. HIIT generally results in higher calorie burn, while Yoga tends to have lower values. However, all workout types include people with low to very high calorie burns.

```python
# Multivariate Analysis using Seaborn
sns.scatterplot(
    data=df,
    x='Session_Duration (hours)',
    y='Calories_Burned',
    hue='Workout_Type',
    size='Burns_Calories_Bin',
    palette='Set1',edgecolor='black')
plt.title("Workout Duration vs Calories Burned with Workout Type and Burn Category")
plt.xlabel("Workout Duration (hours)")
plt.ylabel("Calories Burned")
plt.legend(title="Workout Type")
plt.show()
```

Observation:
- As Workout Duration increases, Calories Burned also increases.
- HIIT and Strength workouts (red, blue) burn more calories even with shorter durations.
- Yoga and Cardio (purple, green) generally show lower calorie burns for the same time.
-The circle sizes increase with longer durations  meaning people who exercise longer also tend to fall in high or very high burn categories.

```python
#Pairplot using seaborn(it helps to see how different columns in this dataset are related to each other.
#Relationship among Calories burned, Session duration and Resting BPM
sns.pairplot(df,vars=['Calories_Burned', 'Session_Duration (hours)', 'Resting_BPM'], hue='Workout_Type', palette='Set2')
plt.suptitle("Pairwise Relationships among Workout Features")
plt.show()
```

Observation:
The diagonal graphs(histogram) are univariate,it shows:
1.Most people burn fewer calories and do shorter workouts, while a few burn a lot.
2.Resting_BPM (heart rate) is more balanced for everyone.(3rd one).
The non diagonol graphs(Scatter plots) are bivariate, it shows:
1.As session duration increases, calories burned also increases.
2. Resting BPM does not change much with other features.
